### Описание датасета:

Name - имя покемона, текстовое значение.

Type 1 - Каждый покемон имеет свой тип, это определяет слабость/устойчивость к атакам. Текстовое значение.

Type 2 - У некоторых покемонов 2 типа. Текстовое значение.

Total - сумма всех характеристик. Показывает на сколько силен покемон. Дискретная величина.

HP - уровень здоровья. Дискретная величина.

Attack - базовый показатель для обычных атак (н-р царапание, удар). Дискретная величина.

Defense - базовый показатель сопротивления обычным атакам. Дискретная величина.

Sp. Atk - базовый показатель для специальных атак (например, огненный взрыв, пузырьковый луч). Дискретная величина.

Sp. Def - базовый показатель сопротивления специальным атакам. Дискретная величина.

Speed - скорость. Определяет какой покемон атакует первым в каждом раунде. Дискретная величина.

Generation - число поколений. Дискретная величина.

Legendary - легендарный покемон или нет. Текстовое значение.
