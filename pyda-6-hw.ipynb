### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 
за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
electricity = pd.read_csv('C:\\Users\\Dmitry\\Documents\\Учеба\\3 Python и математика для анализа данных\\3.6 Основы Pandas\\power.csv')
electricity.head()
```

```python
baltic = electricity[ ((electricity['country'] == 'Latvia') | (electricity['country'] == 'Lithuania') | \
                    (electricity['country'] == 'Estonia')) & ((electricity['category'] == 4) | (electricity['category'] == 12)\
                | (electricity['category'] == 21)) & ((electricity['year'] >= 2005) & (electricity['year'] <= 2010)) \
                    & (electricity['quantity'] >= 0)] 
baltic.head()
```

```python
baltic.shape
```

```python
baltic['quantity'].sum()
```
