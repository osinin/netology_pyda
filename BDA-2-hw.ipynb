```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
```

```python
x, y = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']], df['medv']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_train = pd.DataFrame(x_train, columns = x.columns)
x_test = pd.DataFrame(x_test, columns = x.columns)
```

```python
x_train.head()
```

```python
linear = LinearRegression()
linear.fit(x_train, y_train)
```

```python
svr = SVR()
svr.fit(x_train, y_train)
```

```python
tree = DecisionTreeRegressor()
tree.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# в идеале должно быть 0
mean_squared_error(y_test, linear.predict(x_test))
```

```python
# в идеале должно быть 1
r2_score(y_test, linear.predict(x_test))
```

```python
# в идеале должно быть 0
mean_absolute_error(y_test, linear.predict(x_test))
```

```python
mean_squared_error(y_test, svr.predict(x_test)), r2_score(y_test, svr.predict(x_test)), mean_absolute_error(y_test, svr.predict(x_test))
```

```python
mean_squared_error(y_test, tree.predict(x_test)), r2_score(y_test, tree.predict(x_test)), mean_absolute_error(y_test, tree.predict(x_test))
```

***Обучив модель тремя разными методами и посчитав их качество, можно сделать вывод, что наилучший результат показал метод
DecisionTreeRegressor так как метрики качества ближе к целевым значениям, чем у других методов.***
