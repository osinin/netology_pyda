Таблица цен на дома в Бостоне

Основные значения:

* CRIM: This is the per capita crime rate by town - КРИМИНАЛ
* ZN: This is the proportion of residential land zoned for lots larger than 25,000 sq.ft. - ПРОПОРЦИЯ БОЛЬШИХ УЧАСТКОВ
* INDUS: This is the proportion of non-retail business acres per town.- ДОЛЯ НЕ РОЗНЕЧНОГО БИЗНЕСА
* CHAS: This is the Charles River dummy variable (this is equal to 1 if tract bounds river; 0 otherwise) - У РЕКИ
* NOX: This is the nitric oxides concentration (parts per 10 million) - КОНЦЕНТРАЦИЯ ОКСИДОВ
* RM: This is the average number of rooms per dwelling - СР КОЛИЧЕСТВО КОМНАТ
* AGE: This is the proportion of owner-occupied units built prior to 1940 - ПОСТРОЙКА
* DIS: This is the weighted distances to five Boston employment centers - РАССТОЯНИЕ ДО ЦЕНТРОВ ЗАНЯТОСТИ
* RAD: This is the index of accessibility to radial highways - ИНДЕКС ДОСТУПНОСТИ ТРАСС
* TAX: This is the full-value property-tax rate per 10,000 dollars - НАЛОГ
* PTRATIO: This is the pupil-teacher ratio by town - УЧЕНИК / УЧИТЕЛЬ
* B: This is calculated as 1000(Bk — 0.63)², where Bk is the proportion of people of African American descent by town - ПРОПОРЦИЯ НЕГРОВ
* LSTAT: This is the percentage lower status of the population - ПРОЦЕНТ БЕДНЫХ
* MEDV: This is the median value of owner-occupied homes in 1000 dollars - СРЕДНЯЯ ЦЕНА

**Регрессионный анализ** - можно ли провести линейную зависимость между двумя выборками. Возьмем Ladder - удовлетворение жизнью, и generosity - доброта. Мы хотим найти прямую, которая разделит точки в пространстве наполовину. Это можно сделать графически. Нарисовать график. Можно ли сказать, что данная прямая является собственным вектором? При проецировании данных на эту прямую разброс данных максимален?

```python
ladder = hapiness2['Ladder']
generosity = hapiness2['Generosity']
sns.regplot(x= ladder, y=generosity,data=hapiness2, fit_reg=True)
plt.xlabel('Удовлетворенность жизнью')
plt.ylabel('Доброта')
plt.title('Корреляционная зависимость двух выборок')
plt.show()
# является, максимален
```

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
# Количество испытаний
trials = 10

# Массив значений
values = np.random.randint(1, 7, size=trials)
plt.hist(values, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])

plt.xlabel('Значение')
plt.ylabel('Количества раз')
plt.title('Подбрасывание кубика ' + str(trials) + ' раз')
plt.show()
# подбрасывание кубика - это равномерное распределение, но из-за малого числа испытаний наша гистограмма пока так не выглядит
```

Напишите код для моделирования 1000 подбрасываний игрального кубика. Начертите диаграму histogram. Почему график изменился? Как применить центральную предельную теорему?

```python
# Количество испытаний
trials = 1000

# Массив значений
values = np.random.randint(1, 7, size=trials)
plt.hist(values, bins=[0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])

plt.xlabel('Значение')
plt.ylabel('Количества раз')
plt.title('Подбрасывание кубика ' + str(trials) + ' раз')
plt.show()
# вот теперь распределение гораздо больше напоминает равномерное. чем больше будет экспериментов, тем идеальнее будет
# равномерное распределение.
```

```python
import random
s = np.random.randint(1, 7, size=trials)
samples = [ np.mean(random.choices(s, k=30)) for _ in range(1000) ]
plt.hist(samples)
plt.title('Средние значения выборок размера 30 из исходного распределения')
plt.show()
```
